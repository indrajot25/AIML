```python
import pandas as pd
import numpy as np

# Create a new sample dataset
data = {
    'School ID': [201, 202, 203, 204, 205, 206, 207, 208],
    'Admission No.': [21617, 21432, 21989, 22456, 22001, 21607, 21234, 21876],
    'Name': ['Ritika Sharma', 'Vikram Singh', 'Ravi Verma', 'Simran Kaur', 'Arjun Choudhary', 'Anjali Patel', 'Rajesh Mehra', 'Kavita Das'],
    'Address': ['E-43 CP', 'B-15 RT', 'C-91 LP', 'A-45 TM', 'D-12 VP', 'B-62 MG', 'A-75 KB', 'C-14 TG'],
    'City': ['Delhi', 'Mumbai', 'Chennai', 'Bangalore', 'Pune', 'Hyderabad', 'Ahmedabad', 'Kolkata'],
    'Subject': ['Physics', 'Chemistry', 'Biology', 'Physics', 'Math', 'History', 'English', 'Geography'],
    'Marks': [88, 92, 85, 91, 89, 84, 90, 87],
    'Rank': [2, 1, 4, 3, 5, 6, 7, 8],
    'Grade': ['A+', 'O', 'A', 'O', 'A+', 'B+', 'A', 'B']
}

# Creating Pandas Dataframe
df = pd.DataFrame(data)

print("Updated dataframe:")
df
```

```python
# Fill missing values for categorical columns with the mode (most frequent value)
df.select_dtypes(include='object').fillna(
    df.select_dtypes(include='object').mode().iloc[0]
)
```

```python
# Displaying the first 6 rows of the DataFrame
df.head(6)
```

```python
# Display DataFrame info (columns, data types, non-null counts)
df.info()
```

```python
# Display the count of missing values in each column
df.isna().sum()
```

```python
# Listwise deletion: Remove rows with any missing values
df_listwise = df.dropna()
print("\nDataFrame after listwise deletion:")
 # Displaying DataFrame after removing rows with missing values.
print(df_listwise)
```

```python
# Mean Imputation: Filling missing values in 'Marks' column with the mean
df_mean_imputed = df.copy()  # Create a copy of the original DataFrame
mean_marks = df_mean_imputed['Marks'].mean()  # Calculate mean of 'Marks' column
df_mean_imputed['Marks'].fillna(mean_marks, inplace=True)  # Fill missing 'Marks' values with the mean

print("\nDataFrame after mean imputation:")
print(df_mean_imputed)  # Displaying DataFrame after mean imputation
```

```python
# Outlier Detection: Z-Score method
from scipy import stats
# Calculate Z-scores for the 'Marks' column
z_scores = np.abs(stats.zscore(df['Marks']))
threshold = 80  # Define a threshold for Z-score
outliers = np.where(z_scores > threshold)  # Identify outliers with Z-scores beyond the threshold

print("\nOutliers based on Z-score:")
print(df.iloc[outliers])  # Display the rows that are considered outliers
```

```python
# IQR Method for Outlier Detection
Q1 = df['Marks'].quantile(0.25)  # Calculate the 1st quartile (25th percentile)
Q3 = df['Marks'].quantile(0.75)  # Calculate the 3rd quartile (75th percentile)
IQR = Q3 - Q1  # Calculate the interquartile range (IQR)

# Define lower and upper bounds for detecting outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify rows where 'Marks' is below the lower bound or above the upper bound (outliers)
outliers = df[(df['Marks'] < lower_bound) | (df['Marks'] > upper_bound)]

# Remove the outliers from the DataFrame
df_cleaned = df[(df['Marks'] >= lower_bound) & (df['Marks'] <= upper_bound)]

print("\nDataFrame after removing outliers (IQR method):")
print(df_cleaned)  # Displaying the DataFrame after outlier removal using the IQR method
```
